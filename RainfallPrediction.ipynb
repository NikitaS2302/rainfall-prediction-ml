```python
#importing libraries
import numpy as np
import pandas as pd
import sklearn as sk
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
rain=pd.read_csv('austin_weather_final.csv')
rain
```

```python
X=rain.drop(['PrecipitationSumInches'],axis=1)
```

```python
Y=rain['PrecipitationSumInches']
#reshaping into 2d vector
Y=Y.values.reshape(-1,1)
Y
```

```python
day_index=500
day=[i for i in range(Y.size)]
```

```python
#intiialise the Linear Regression classifier
clf=LinearRegression()
clf.fit(X,Y)
```

```python
#plotting graph
plt.scatter(day,Y,color='blue')
plt.scatter(day[day_index],Y[day_index],color='red')
plt.title('Precipation trend graph',fontdict={'fontsize':24,'fontweight':150,'color':'brown'})
plt.xlabel('Days')
plt.ylabel('Precipatation_in_inches')
plt.show()
x_vis=X.filter(['T,empAvgF','DewPointAvgF','HumidityAvgPercent','SeaLevelPressureAvgInches','VisibilityAvgMiles','WindAvgMPH'])
```

```python
plt.title('Precipitaion vs some attributes trend graph\n')
for i in range(x_vis.columns.size):
    plt.subplot(3,2,i+1)
    plt.scatter(day,x_vis[x_vis.columns.values[i][:100]],color='green')
    plt.scatter(day[day_index],x_vis[x_vis.columns.values[i]][day_index],color='cyan')
    plt.title(x_vis.columns.values[i])
plt.tight_layout()
plt.show()
```
